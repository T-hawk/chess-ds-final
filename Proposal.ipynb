# Background and Motivation
The game of chess is an incredibly fascinating, yet deceptively simple, game. With only six unique pieces, each constrained to a number of legal moves on a small 8x8 grid, the game may seem trivial. Yet, chess remains infinitely complex, boasting an estimated 10^120 unique games. This extensive replayability and strategy behind the game has allowed chess to withstand the test of time, being played for over a thousand years all the way into the digital age. And this is where our project motivation stems from: chess in the modern day. Because the game is so popular, there are multiple robust datasets of online chess games publicly available. These datasets provide everything from the game winner, to the moves that were made, to the rating (ELO) of the players. Having this much accessible data would allow for extensive statistical analysis, revealing new trends in the digital era or retention of old reliable strategies to this day. Additionally, the implementation of chess engines, such as Stockfish, may prove interesting in effectively analyzing strategies and game state evaluations. We aim to study a dataset of games from the public chess platform Lichess in the hopes of discovering such patterns.


# Project Objectives

- What strategies/openings are most effective overall?
- What strategies/openings are most effective by ranking?

# Data Description and Acquisition
The game data will be obtained from Lichess, which is a free and open source chess server. Around 5 million games of chess are played every day on the site, with several publicly available datasets readily accessible online. Here is a breakdown of our prospective data sources:
- Kaggle database: https://www.kaggle.com/datasets/datasnaek/chess/data
  - Single downloadable CSV
  - Nice, relatively clean format (will serve as initial dataset)
  - 20,058 rows
  - 16 Columns: Game ID, Rated (T/F), Start Time, End Time, Number of Turns, Game Status, Winner, Time Increment, White Player ID, White Player Rating, Black Player ID, Black Player Rating, All Moves in Standard Chess Notation, Opening ECO (standardised code for any given opening), Opening Name, and Opening Ply (number of moves in the opening phase)
- Directly from Lichess: https://database.lichess.org/#standard_games
  - PGN (portable game notation) files for every month since 2013
  - Requires more data cleaning and processing
  - Massive 30GB datasets for every month (around 90,000,000 games per month)
  - PGN contains Winner, Player ELOs, Date, Opening, and a List of All Moves

Other considerations not mentioned above:
- This data may not represent the entire chess community as a whole because it is from Lichess.org. The most popular chess website is Chess.com, so our dataset does not fully represent the online chess community. Additionally, there are many chess players that play in-person over the board (OTB) instead of online, which further divorces our dataset from the broader chess community. This could be potentially harmful if we use certain language in the project, such as “the chess community” instead of “the Lichess community,” as this would misrepresent the chess community while we are only able to study a small subset of it. Such risks can be reduced by acknowledging the appropriate scale of our project at each stage of analysis.
- This project could potentially be used for harm if malicious actors presented our model to convince vulnerable people that their chess skills are inadequate. This could occur in multiple ways, such as advertising our model as a sales tactic to promote counterfeit chess courses or abusing the model’s ELO prediction as a form of cyber bullying. Such risks can be reduced by limiting or monitoring distribution of public access to our model.


# Data Cleaning and Processing
- The Kaggle dataset linked above will serve as our primary dataset, but larger datasets may be added as the project develops. The following steps will be taken for each dataset:
- Data cleaning will consist of the following:
  - Assigning categorical numeric data for rated (T/F), victory status, winner, and opening ECO
  - Ensure all numeric data is correctly assigned to the correct type
  - Convert created at and last move at into datetime
  - Create new column for total time taken (using created at and last move at columns)
  - There are no missing data points according to Kaggle, so no None conversions are necessary
- Data processing will be a heavy task as it will involve move analysis. We will need to develop algorithms and methods for analyzing the list of moves. This will likely involve pulling in Stockfish (the chess engine) to help with board analysis.
